```python
import numpy as np
```

```python
# Level 1: Creating Numpy Arrays

a=np.array([1,5,6,9,8])
print(a)
```

```python
# Level 2: Arrays with Zeros And Ones

Zero=np.zeros(4)
one=np.ones((2,3))

print(Zero)
print()
print(one)
```

```python
# Level 3: Random Number Arrays

random=np.random.rand(3,2)
print(random)
```

```python
# Level 4: Reshaping Arrays

a=np.array([1,2,5,8,9,6])
print(a.reshape(3,2))
```

```python
# Level 5: Array Indexing and Slicing

a=np.array([1,2,5,8,6])
print(a[2])
print(a[1:5])
```

```python
# Level 6: basic mathematical Operations

a=np.array([1,5,8])
b=np.array([8,4,5])
r=a+b
print(r)
```

```python
# Level 7: Statistical Operations

a=np.array([10,20,50,40,60])
print(np.mean(a))
print(np.median(a))
print(np.std(a))
```

```python
# Level 8: Broadcasting

a=np.array([1,5,6])
r=a+5
print(r)
```

```python
# Level 9:Boolean Indexing

a=np.array([12,15,18,16,23,24])
f=a[a>20]
print(f)
```

```python
# Level 10: Vectorized Operations

a=np.array([1,2,3,4,5])
r=np.sqrt(a)
print(r)
```

```python
# Level 11: Liner Algebra Operations

a=np.array([[1,2],[3,4]])
b=np.array([[2,0],[1,2]])
product=np.dot(a,b)
determinant=np.linalg.det(a)
print(product)
print(determinant)
```

```python
# Level 12: Aggregation Functions

a=np.array([3,6,9,4])
print(np.sum(a))
print(np.min(a))
print(np.max(a))
print(np.cumsum(a))
```

```python
# Level 13: Sorting and Searching

a=np.array([15,22,25,86,6])
sort=np.sort(a)
index=np.where(a>25)
print(sort)
print(index)
```

```python
# Level 14: Concatenation and Splitting

x=np.array([1,5,6,9])
y=np.array([8,6,3,2])
combined=np.concatenate((x,y))
split=np.split(combined,2)
print(combined)
print(split)
```
